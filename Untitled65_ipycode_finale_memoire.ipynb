```python
# ============================================================
# SECURE LOG MANAGEMENT FRAMEWORK
# SHA256 + MERKLE TREE + BLOCKCHAIN + DEEPLOG
# ============================================================

import pandas as pd
import numpy as np
import hashlib
import json
import time
from datetime import datetime

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.preprocessing import LabelEncoder

# ============================================================
# 1. LOAD LOGS
# ============================================================

PATH = "/content/drive/MyDrive/HDFS_2k.log_structured.csv"

df = pd.read_csv(PATH)

print("Dataset:", df.shape)

# ============================================================
# 2. SHA256 HASHING
# ============================================================

def sha256(data):
    return hashlib.sha256(
        str(data).encode()
    ).hexdigest()

df["hash"] = df["EventTemplate"].astype(str).apply(
    sha256
)

print("\nHashing completed")

# ============================================================
# 3. MERKLE TREE
# ============================================================

def merkle_root(hashes):

    hashes = list(hashes)

    while len(hashes) > 1:

        new_hashes = []

        for i in range(0, len(hashes), 2):

            left = hashes[i]

            if i + 1 < len(hashes):
                right = hashes[i + 1]
            else:
                right = left

            new_hash = sha256(left + right)

            new_hashes.append(new_hash)

        hashes = new_hashes

    return hashes[0]

root = merkle_root(df["hash"])

print("\nMerkle Root:")
print(root)

# ============================================================
# 4. SIMPLE PRIVATE BLOCKCHAIN
# ============================================================

class Block:

    def __init__(
        self,
        index,
        timestamp,
        data,
        previous_hash
    ):

        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash

        self.hash = self.compute_hash()

    def compute_hash(self):

        block_string = json.dumps(
            {
                "index": self.index,
                "timestamp": self.timestamp,
                "data": self.data,
                "previous_hash": self.previous_hash
            },
            sort_keys=True
        )

        return sha256(block_string)

class Blockchain:

    def __init__(self):

        self.chain = []

        self.create_genesis_block()

    def create_genesis_block(self):

        genesis = Block(
            0,
            str(datetime.now()),
            "GENESIS",
            "0"
        )

        self.chain.append(genesis)

    def add_block(self, data):

        previous = self.chain[-1]

        block = Block(
            len(self.chain),
            str(datetime.now()),
            data,
            previous.hash
        )

        self.chain.append(block)

blockchain = Blockchain()

blockchain.add_block(
    {
        "MerkleRoot": root
    }
)

print("\nBlockchain Length:", len(blockchain.chain))

# ============================================================
# 5. DEEPLOG PREPARATION
# ============================================================

encoder = LabelEncoder()

df["event_id"] = encoder.fit_transform(
    df["EventId"].astype(str)
)

events = df["event_id"].values

SEQ_LEN = 10

X = []
y = []

for i in range(len(events)-SEQ_LEN):

    X.append(events[i:i+SEQ_LEN])

    y.append(events[i+SEQ_LEN])

X = np.array(X)
y = np.array(y)

split = int(0.7 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

# ============================================================
# 6. DEEPLOG MODEL
# ============================================================

num_classes = len(encoder.classes_)

model = Sequential([

    Embedding(
        num_classes,
        128
    ),

    LSTM(
        128,
        return_sequences=True
    ),

    Dropout(0.2),

    LSTM(
        64
    ),

    Dense(
        num_classes,
        activation="softmax"
    )
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# ============================================================
# 7. TRAINING
# ============================================================

start = time.time()

history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_data=(X_test, y_test),
    callbacks=[
        EarlyStopping(
            patience=3,
            restore_best_weights=True
        )
    ],
    verbose=1
)

train_time = time.time() - start

# ============================================================
# 8. DEEPLOG ANOMALY DETECTION
# ============================================================

proba = model.predict(
    X_test,
    verbose=0
)

TOP_K = 5

topk = np.argsort(
    proba,
    axis=1
)[:, -TOP_K:]

anomaly_pred = np.array([
    0 if y_test[i] in topk[i]
    else 1
    for i in range(len(y_test))
])

anomaly_rate = anomaly_pred.mean()

top1_acc = np.mean(
    np.argmax(proba, axis=1) == y_test
)

top5_acc = np.mean([
    y_test[i] in topk[i]
    for i in range(len(y_test))
])

# ============================================================
# 9. TAMPERING SIMULATION
# ============================================================

original_log = str(
    df["EventTemplate"].iloc[0]
)

tampered_log = original_log + "_MODIFIED"

original_hash = sha256(
    original_log
)

tampered_hash = sha256(
    tampered_log
)

integrity_status = (
    original_hash == tampered_hash
)

# ============================================================
# 10. FINAL REPORT
# ============================================================

print("\n==============================")
print("SECURE LOG FRAMEWORK RESULTS")
print("==============================")

print("\nTraining Time:",
      round(train_time,2),"sec")

print("Top-1 Accuracy:",
      round(top1_acc,4))

print("Top-5 Accuracy:",
      round(top5_acc,4))

print("Anomaly Rate:",
      round(anomaly_rate,4))

print("\nBlockchain Blocks:",
      len(blockchain.chain))

print("\nMerkle Root:")
print(root[:50], "...")

print("\nTampering Detection:")
print("Integrity Preserved:",
      integrity_status)

print("==============================")
```
